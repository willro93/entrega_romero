# Información nutricional de alimentos enfocados en la creación de dietas. 
El valor nutricional de los alimentos incluye propiedades nutricionales como lípidos, minerales, vitaminas, oligoelementos y carbohidratos. Por lo tanto, se puede determinar que el valor nutricional es sinónimo del valor nutricional de los alimentos. Sobre esta base, los alimentos contribuyen significativamente al aporte total que resulta necesario para una dieta sana y balanceada.
La importancia de realizar una dieta balanceada resulta cada vez más importante debido al aumento de enfermedades relacionadas con una mala alimentación tales como diabetes, anemia, caries, obesidad, enfermedades cardiovaasculares, entre otras, por que lo que es de interes estudiar los nutrientes de los alimentos.
Los nutrientes que se encuentran en los alimentos se dividen en macronutrientes como lo son: proteínas, carbohidratos y grasas; y los micronutrientes los cuales se componen por: vitaminas y minerales, los cuales resultan indispensables en cantidades pequeñas para un desarrollo. Conocer la información nutricional de los alimentos ayuda a la creación de mejores dietas basadas en el control de porciones y las necesidades nutricionales de pacientes que deseen hacer una reducción o aumento de de peso, prevenir enfermedades, controlar enfermedades o reducir el consumo de ciertos minerales.
Un estudio de manera individual y grupal permite conocer las cantidades adecuadas de consumo para cada alimento, y encontrar una relación entre ellos y la modificación de cantidades ya sea de macronutriewntes y micronutrientes. El porcentaje diario recomendado para cada alimento es una guía fundamental que proporciona cotas inferiores y superiores que aunado a las relaciones encontradas en un estudio realizado a un base de datos permite ofrecer mejores alternativas de consumo.


# Preguntas de interés e hipótesis.
Se realizará un análisis de los nutrientes contenidos en los alimentos con el objetivo de encontrar relaciones entre las distintas variables que existen por ejemplo ¿qué relación existe entre el contenido calórico (kcal) y el contenido de grasa (g)?, y basado en dichas relaciones ¿cómo proporcionar mejores alternarivas de alimentación?.

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df_ingredientes= pd.read_csv("cleaned_ingredients.csv", sep=",")
df_ingredientes
```

```python
graf_disp=sns.scatterplot(data=df_ingredientes, x="Energy_kcal", y="Fat_g")
graf_disp.set_xlabel("Energía (kcal)")
graf_disp.set_ylabel("Grasa (g)")
graf_disp.set_title("Relación entre la energía proporcionada contra contenido de grasa")
```

**El siguiente gráfico presenta la relación que existe entre los alimentos con grasas y la cantidad de kcal que aporta, se puede apreciar una relación entre el aumento de grasa y el aporte calórico.**

```python
graf_linea=sns.scatterplot(data=df_ingredientes, x="Energy_kcal", y="Fiber_g")
graf_linea.set_xlabel("Energía (kcal)")
graf_linea.set_ylabel("Fibra (g)")
graf_linea.set_title("Relación entre la energía proporcionada contra contenido de fibra")
```

**En la siguiente gráfica se observa que aporte calórico no tiene una relación directa con el aporte de fibra de los alimentos**

```python
graph_B12= sns.scatterplot(data=df_ingredientes, x="VitB12_mcg", y="Saturated_fats_g")
graph_B12.set_xlabel("Vitamina B12 (mg)")
graph_B12.set_ylabel("Grasas saturadas (g)")
graph_B12.set_title("Relación entre el contenido de grasas saturadas contra el contenido de vitamina B12")
```

**En la siguiente gráfica se aprecia La relación que existe entre los alimentos con mayor contenido de grasas saturadas y vitamina B12, los alimentos con mayor cantida de grasas saturadas poseen cantidades menores de vitamina B12**

**A partir de la promeras gráficas presentadas se puede concluir que con el objetivo de generar dietas más ricas en nutrientes y bajas en grasas, se observa que los alimentos que contienen más calorías tienen una relación con lo alimentos que más grasa contiene, de manera similar los alimentos que contiene mayor cantidad de grasas saturadas contienen menos cantidades de vitamina B12. Por otro lado los alimentos que contienen ciertas cantidades de fibra, no guardan una relación con las calorías contenidas en dichos alimentos**
